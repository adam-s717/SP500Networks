```python
import datetime as dt
import numpy as np
import os
import pandas as pd
import pandas_datareader as web
```

```python
AllReturns = pd.read_csv('sp500_joined_returns.csv')
OhOh_OhEight = AllReturns[(AllReturns['Date'] > '2000-01-01') & (AllReturns['Date'] <= '2007-12-31')]
OhEight_Ten = AllReturns[(AllReturns['Date'] > '2008-01-01') & (AllReturns['Date'] <= '2009-12-31')]
Ten_Fifteen = AllReturns[(AllReturns['Date'] > '2010-01-01') & (AllReturns['Date'] <= '2014-12-31')]
Fifteen_Current = AllReturns[(AllReturns['Date'] > '2015-01-01') & (AllReturns['Date'] <= '2021-12-31')]
Twenty_TwentyOne = AllReturns[(AllReturns['Date'] > '2020-01-01') & (AllReturns['Date'] <= '2021-12-31')]
```

```python
Twenty_TwentyOne
```

```python
W_AR = AllReturns.corr(method='spearman').fillna(0)
W_OE = OhOh_OhEight.corr(method='spearman').fillna(0)
W_ET = OhEight_Ten.corr(method='spearman').fillna(0)
W_TF = Ten_Fifteen.corr(method='spearman').fillna(0)
W_FC = Fifteen_Current.corr(method='spearman').fillna(0)
W_TwTo = Twenty_TwentyOne.corr(method='spearman').fillna(0)
```

```python
if not os.path.exists('Weighted_Adjacency'):
    os.makedirs('Weighted_Adjacency')

W_AR.to_csv('Weighted_Adjacency/W_AR.csv')
W_OE.to_csv('Weighted_Adjacency/W_OE.csv') 
W_ET.to_csv('Weighted_Adjacency/W_ET.csv') 
W_TF.to_csv('Weighted_Adjacency/W_TF.csv') 
W_FC.to_csv('Weighted_Adjacency/W_FC.csv') 
W_TwTo.to_csv('Weighted_Adjacency/W_TwTo.csv')


```
