```python
import datetime as dt
import numpy as np
import os
import pandas as pd
import pandas_datareader as web
```

```python
W_AR = pd.read_csv('Weighted_Adjacency/W_AR.csv',index_col=0)
W_OE = pd.read_csv('Weighted_Adjacency/W_OE.csv',index_col=0)
W_ET = pd.read_csv('Weighted_Adjacency/W_ET.csv',index_col=0)
W_TF = pd.read_csv('Weighted_Adjacency/W_TF.csv',index_col=0)
W_FC = pd.read_csv('Weighted_Adjacency/W_FC.csv',index_col=0) 
W_TwTo = pd.read_csv('Weighted_Adjacency/W_TwTo.csv',index_col=0)
```

```python
W_OE
```

```python
W_AR.values[[np.arange(W_AR.shape[0])]*2] = 1 #Forcing the diagonal to be 1 since we set all NaN's to 0
W_AR.where(W_AR <= .6, 1,inplace=True) #If its greater than .6 replace with 1
W_AR.where(W_AR == 1, 0,inplace=True)  #If it doest equal 1 replace with 0
A_AR = W_AR

W_OE.values[[np.arange(W_OE.shape[0])]*2] = 1 #Forcing the diagonal to be 1 since we set all NaN's to 0
W_OE.where(W_OE <= .6, 1,inplace=True)
W_OE.where(W_OE == 1, 0,inplace=True)
A_OE = W_OE

W_ET.values[[np.arange(W_ET.shape[0])]*2] = 1 #Forcing the diagonal to be 1 since we set all NaN's to 0
W_ET.where(W_ET <= .6, 1,inplace=True)
W_ET.where(W_ET == 1, 0,inplace=True)
A_ET = W_ET

W_TF.values[[np.arange(W_TF.shape[0])]*2] = 1 #Forcing the diagonal to be 1 since we set all NaN's to 0
W_TF.where(W_TF <= .6, 1,inplace=True)
W_TF.where(W_TF == 1, 0,inplace=True)
A_TF = W_TF

W_FC.values[[np.arange(W_FC.shape[0])]*2] = 1 #Forcing the diagonal to be 1 since we set all NaN's to 0
W_FC.where(W_FC <= .6, 1,inplace=True)
W_FC.where(W_FC == 1, 0,inplace=True)
A_FC = W_FC

W_TwTo.values[[np.arange(W_TwTo.shape[0])]*2] = 1 #Forcing the diagonal to be 1 since we set all NaN's to 0
W_TwTo.where(W_TwTo <= .6, 1,inplace=True)
W_TwTo.where(W_TwTo == 1, 0,inplace=True)
A_TwTo = W_TwTo
```

```python
if not os.path.exists('Adjacency'):
    os.makedirs('Adjacency')

A_AR.to_csv('Adjacency/A_AR.csv')
A_OE.to_csv('Adjacency/A_OE.csv') 
A_ET.to_csv('Adjacency/A_ET.csv') 
A_TF.to_csv('Adjacency/A_TF.csv') 
A_FC.to_csv('Adjacency/A_FC.csv') 
A_TwTo.to_csv('Adjacency/A_TwTo.csv')

```
